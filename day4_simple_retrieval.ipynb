```python
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = [
    # Machine Learning
    "Transformers use self-attention mechanisms for sequence modeling in NLP.",
    "Diffusion models generate images by gradually removing noise from random data.",
    "Embeddings convert text into dense vectors that represent semantic meaning.",
    # Cooking
    "Pasta carbonara is made with eggs, cheese, pancetta, and black pepper.",
    "Sourdough bread requires a live starter culture for fermentation.",
    # Space
    "The James Webb Space Telescope captures infrared images of distant galaxies.",
    "Mars has the largest volcano in the solar system, Olympus Mons.",
    # Programming
    "Python is a high-level language known for its readable and concise syntax.",
    "Git is a distributed version control system for tracking code changes.",
    # Sports
    "The FIFA World Cup is held every four years and is the largest football tournament.",
    "Michael Jordan won six NBA championships with the Chicago Bulls.",
]
```

```python

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
doc_embedding = model.encode(documents)
```

```python
doc_embedding.shape
```

```python
query = "How do transformers model sequences in NLP?"
query_embedding = model.encode(query)
```

```python
score = cosine_similarity(query_embedding.reshape(1, -1), doc_embedding)
```

```python
print(score)
```

```python
score = score.flatten()
```

```python
print(score.shape)
```

```python
k = 2
ranked_indices = np.argsort(score)[::-1][:2]
```

```python
print(ranked_indices)
```

```python
top_k_docs = [documents[i] for i in ranked_indices]
```

```python
print(top_k_docs)
```
